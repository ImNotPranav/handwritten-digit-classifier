```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.activations import linear, relu, sigmoid
mnist=tf.keras.datasets.mnist
(x_train, y_train),(x_test, y_test)=mnist.load_data()
model= Sequential([
    Input(shape=(28,28)),
    Flatten(),
    Dense(units=70, activation='relu'),
    Dense(units=35, activation='relu'),
    Dense(units=10, activation='linear')
])
model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(0.001)
)
model.fit(
    x_train,y_train, epochs=40
)
```

```python
logits = model.predict(x_test)
predicted_classes = np.argmax(logits, axis=1)
for i in range(10):
    print(f"Image {i}: Predicted = {predicted_classes[i]}, Actual = {y_test[i]}")
```

```python
import matplotlib.pyplot as plt

for i in range(5):
    plt.imshow(x_test[i], cmap='BuPu')
    plt.title(f"Predicted: {predicted_classes[i]}, Actual: {y_test[i]}")
    plt.axis('off')
    plt.show()
```

```python
wrong_indices = np.where(predicted_classes != y_test)[0]

print(f"Total wrong predictions: {len(wrong_indices)}")
for i in wrong_indices[:5]:
    print(f"Index {i}: Predicted = {predicted_classes[i]}, Actual = {y_test[i]}")
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Wrong Prediction")
    plt.axis('off')
    plt.show()
```
